```python
import numpy as np
from scipy.stats import norm
np.random.seed(42)
population = np.random.randint(50, 96, 1000)
sample_size = int(0.10 * len(population))
sample = np.random.choice(population, size=sample_size, replace=False)
population_mean = np.mean(population)
population_std = np.std(population)
sample_mean = np.mean(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))
print("Population Size:", len(population))
print("Sample Size:", sample_size)
print("Population Mean:", population_mean)
print("Population Standard Deviation:", population_std)
print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)
alpha = 0.05
if p_value < alpha:
    print("Decision: Reject the Null Hypothesis")
else:
    print("Decision: Fail to Reject the Null Hypothesis")
```

```python
import numpy as np
from scipy import stats
from statsmodels.stats.weightstats import ztest
print("----- Z-Test -----")
sample_marks = np.array([78, 82, 85, 88, 90, 76, 95, 89, 84, 91])
population_mean = 80
z_stat, p_value = ztest(sample_marks, value=population_mean)
print("Sample Marks:")
print(sample_marks)
print("\nPopulation Mean:", population_mean)
print("\nZ-Statistic:", z_stat)
print("P-Value:", p_value)
alpha = 0.05
if p_value < alpha:
    print("\nResult: Reject Null Hypothesis")
else:
    print("\nResult: Fail to Reject Null Hypothesis")
print("\n\n----- Independent t-Test -----")
class_A = np.array([78, 82, 85, 88, 90])
class_B = np.array([72, 75, 70, 68, 74])
t_stat, p_value = stats.ttest_ind(class_A, class_B)
print("Class A Marks:")
print(class_A)
print("\nClass B Marks:")
print(class_B)
print("\nt-Statistic:", t_stat)
print("P-Value:", p_value)
if p_value < alpha:
    print("\nResult: Reject Null Hypothesis")
else:
    print("\nResult: Fail to Reject Null Hypothesis")
```

```python
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.weightstats import ztest
from statsmodels.stats.proportion import proportions_ztest
np.random.seed(42)
group1 = np.random.normal(loc=70, scale=10, size=30)
group2 = np.random.normal(loc=75, scale=10, size=30)
group3 = np.random.normal(loc=80, scale=10, size=30)
before = np.random.normal(loc=65, scale=8, size=30)
after = before + np.random.normal(loc=5, scale=4, size=30)
z_stat, p_value = ztest(group1, value=70)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
z_stat, p_value = ztest(group1, group2)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
success_count = 60
sample_size = 100
claimed_proportion = 0.50
z_stat, p_value = proportions_ztest(
    count=success_count,
    nobs=sample_size,
    value=claimed_proportion
)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
successes = np.array([60, 75])
samples = np.array([100, 120])
z_stat, p_value = proportions_ztest(
    count=successes,
    nobs=samples
)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
t_stat, p_value = stats.ttest_1samp(group1, popmean=70)
print("T-statistic:", t_stat)
print("P-value:", p_value)
t_stat, p_value = stats.ttest_ind(group1, group2)
print("T-statistic:", t_stat)
print("P-value:", p_value)
t_stat, p_value = stats.ttest_rel(before, after)
print("T-statistic:", t_stat)
print("P-value:", p_value)
f_stat, p_value = stats.f_oneway(group1, group2, group3)
print("F-statistic:", f_stat)
print("P-value:", p_value)
x = np.random.normal(50, 10, 30)
y = x + np.random.normal(5, 5, 30)
r_value, p_value = stats.pearsonr(x, y)
print("Correlation Coefficient:", r_value)
print("P-value:", p_value)
slope, intercept, r_value, p_value, std_error = stats.linregress(x, y)
print("Slope:", slope)
print("Intercept:", intercept)
print("R-value:", r_value)
print("P-value:", p_value)
print("Standard Error:", std_error)
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)
f_stat = var1 / var2
df1 = len(group1) - 1
df2 = len(group2) - 1
p_value = 2 * min(
    stats.f.cdf(f_stat, df1, df2),
    1 - stats.f.cdf(f_stat, df1, df2)
)
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
import numpy as np
from scipy import stats
np.random.seed(42)
group1 = np.array([45, 50, 52, 55, 58, 60, 62, 65])
group2 = np.array([48, 53, 57, 59, 61, 64, 67, 70])
group3 = np.array([40, 44, 49, 51, 54, 56, 59, 63])
before = np.array([50, 52, 55, 57, 60, 62, 64, 66])
after  = np.array([53, 55, 56, 60, 63, 65, 67, 70])
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([2, 4, 5, 7, 8, 10, 11, 13])
observed = np.array([25, 30, 45])
expected = np.array([33.33, 33.33, 33.34])
u_stat, p_value = stats.mannwhitneyu(group1, group2)
print("U-statistic:", u_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
w_stat, p_value = stats.wilcoxon(before, after)
print("W-statistic:", w_stat)
print("P-value:", p_value)
print("\n========== Kruskal-Wallis H Test ==========")
h_stat, p_value = stats.kruskal(group1, group2, group3)
print("H-statistic:", h_stat)
print("P-value:", p_value)
print("\n========== Friedman Test ==========")
condition1 = np.array([55, 60, 62, 64, 66])
condition2 = np.array([58, 61, 65, 67, 69])
condition3 = np.array([60, 63, 66, 70, 72])
friedman_stat, p_value = stats.friedmanchisquare(
    condition1, condition2, condition3
)
print("Friedman Statistic:", friedman_stat)
print("P-value:", p_value)
rho, p_value = stats.spearmanr(x, y)
print("Spearman Correlation:", rho)
print("P-value:", p_value)
tau, p_value = stats.kendalltau(x, y)
print("Kendall Tau:", tau)
print("P-value:", p_value)
print("\n========== Chi-Square Goodness of Fit Test ==========")
chi_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)
print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)
table = np.array([
    [30, 10],
    [20, 40]
])
chi_stat, p_value, dof, expected_freq = stats.chi2_contingency(table)
print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected_freq)
sample = np.random.uniform(0, 1, 100)
ks_stat, p_value = stats.kstest(sample, 'uniform')
print("KS Statistic:", ks_stat)
print("P-value:", p_value)
stat, p_value, median, contingency_table = stats.median_test(
    group1, group2, group3
)
print("Statistic:", stat)
print("P-value:", p_value)
print("Grand Median:", median)
print("Contingency Table:")
print(contingency_table)
differences = after - before
positive_signs = np.sum(differences > 0)
negative_signs = np.sum(differences < 0)
n = positive_signs + negative_signs
p_value = stats.binomtest(
    positive_signs,
    n,
    p=0.5,
    alternative='two-sided'
).pvalue
print("Positive Signs:", positive_signs)
print("Negative Signs:", negative_signs)
print("P-value:", p_value)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)
X_train = np.array([
    [2, 50],
    [3, 60],
    [4, 65],
    [5, 70],
    [6, 75],
    [7, 80],
    [8, 85],
    [9, 90]
])
y_train = np.array([0, 0, 0, 1, 1, 1, 1, 1])
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
X_test = np.array([
    [3, 55],
    [5, 72],
    [8, 88],
    [4, 60]
])
y_test = np.array([0, 1, 1, 0])
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
error_rate = 1 - accuracy
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print("Actual Labels:")
print(y_test)
print("\nPredicted Labels:")
print(y_pred)
print("\nConfusion Matrix:")
print(cm)
print("\nAccuracy:")
print(accuracy)
print("\nError Rate:")
print(error_rate)
print("\nPrecision:")
print(precision)
print("\nRecall:")
print(recall)
print("\nF1-Score:")
print(f1)
```

```python
import pandas as pd
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')
nltk.download('stopwords')
print("----- Text Preprocessing -----")
text = "The movie was amazing, exciting, and full of action scenes!"
text = text.lower()
text = text.translate(str.maketrans('', '', string.punctuation))
tokens = word_tokenize(text)
stop_words = set(stopwords.words('english'))
filtered_tokens = [
    word for word in tokens
    if word not in stop_words
]
print("Original Text:")
print("The movie was amazing, exciting, and full of action scenes!")
print("\nProcessed Tokens:")
print(filtered_tokens)
print("\n\n----- Content-Based Recommendation -----")
movies = pd.DataFrame({
    'Movie': [
        'Avengers',
        'Batman',
        'Superman',
        'Titanic',
        'Inception'
    ],
    'Description': [
        'action superhero marvel fight',
        'dark superhero action detective',
        'superhero flying action hero',
        'romance ship love drama',
        'dream science fiction thriller'
    ]
})
print("Movie Dataset:")
print(movies)
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(movies['Description'])
similarity = cosine_similarity(tfidf_matrix)
movie_name = 'Batman'
movie_index = movies[movies['Movie'] == movie_name].index[0]
scores = list(enumerate(similarity[movie_index]))
sorted_scores = sorted(
    scores,
    key=lambda x: x[1],
    reverse=True
)
print("\nRecommended Movies for Batman:")
for i in sorted_scores[1:4]:
    print(movies.iloc[i[0]]['Movie'])
```
